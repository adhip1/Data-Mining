```python
import pandas as pd
import numpy as np
from scipy.io import arff
from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import random
```

```python
drive.mount('/content/drive')


file_path = '/content/drive/MyDrive/Dry_Bean_Dataset.arff'

data, meta = arff.loadarff(file_path)
df = pd.DataFrame(data)
df
```

```python
unique_values = df['Class'].unique()
print(unique_values)
```

```python
mapping = {value: index for index, value in enumerate(unique_values)}
df['Class'] = df['Class'].map(mapping)
df
```

```python
for i in range (0,17):
  data_type = df.iloc[:, i].dtype
  print(data_type)
```

```python
x = df.iloc[:,0:16]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x.describe()
```

```python
scaler = StandardScaler()
scaler.fit(x)
x_scaled = scaler.transform(x)
x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled_df
```

```python
x_scaled_df.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=43)
```

```python
neigh = KNeighborsClassifier()
neigh.fit(x_train, y_train)
```

```python
y_pred = neigh.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
correlation_matrix = x_scaled_df.corrwith(y)

correlation_df = pd.DataFrame({'Attribute': correlation_matrix.index, 'Correlation': correlation_matrix.values})
correlation_df['Correlation'] = correlation_df['Correlation']

plt.figure(figsize=(10, 6))
plt.bar(correlation_df['Attribute'], correlation_df['Correlation'])
plt.xlabel('Attribute')
plt.ylabel('Correlation')
plt.title('Correlation Values of Attributes with Class Variable')
plt.xticks(rotation=90)
plt.show()
```
